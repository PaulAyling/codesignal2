```python
import copy
def spiralNumbers(n):
    def display_matrix(matrix,n):
        for item in range(n):
            print(matrix[item])
        return
    
    def create_matrix(n):
        for item in range(n):
            row_list.append(0)
        for item in range(n):
            matrix.append(copy.deepcopy(row_list))
        return matrix
    
    def move_right(matrix,current_state):
        row = current_state[0]
        col = current_state[1]
        counter = current_state[2]
        while col < n-1 and matrix[row][col+1] <1:
            col += 1
            matrix[row][col] = counter
            counter += 1
        current_state = [row,col,counter]
        return(current_state)
    
    def move_down(matrix,current_state):
        row = current_state[0]
        col = current_state[1]
        counter = current_state[2]
        while row < n-1 and matrix[row+1][col] <1:
            row += 1
            matrix[row][col] = counter
            counter += 1
        current_state = [row,col,counter]
        return(current_state)
    
    def move_left(matrix,current_state):
        row = current_state[0]
        col = current_state[1]
        counter = current_state[2]
        while col > 0 and matrix[row][col-1] <1:
            col -= 1
            matrix[row][col] = counter
            counter += 1
        current_state = [row,col,counter]
        return(current_state)
    
    def move_up(matrix,current_state):
        row = current_state[0]
        col = current_state[1]
        counter = current_state[2]
        while row > 0 and matrix[row-1][col] <1:
            row -= 1
            matrix[row][col] = counter
            counter += 1
        current_state = [row,col,counter]
        return(current_state)
    
    
    
    row_list = []
    matrix = []
    matrix = create_matrix(n)
    
    row = 0
    col = 0
    counter = 2
    current_state = [row,col,counter]
    new_state = current_state
    matrix[row][col] = 1
    
    print(new_state)
    while new_state[2] <= (n*n):
        
        new_state = move_right(matrix,new_state)
        print("new state",new_state)    
        new_state = move_down(matrix,new_state)
        print("new state",new_state)
        new_state = move_left(matrix,new_state)
        print("new state",new_state)
        new_state = move_up(matrix,new_state)
        print("new state",new_state)
    
    
    display_matrix(matrix,n)
    
    return matrix
```
